```python
#função para preencher os dados das peças
def preenche_peca(pecadiferente, peca,qtdpeca,motivopeca,avaria):
    print("iniciou a função")
    elemento_base = 44
    for inter in range(pecadiferente):
        driver.find_element(By.XPATH ,f'/html/body/div[1]/div/component-dialog-container/component-dialog/div/component-card/div/form/div[1]/div[{elemento_base}]/div[1]/input').click()
        
        time.sleep(5)
        buttons = driver.find_elements(By.CSS_SELECTOR, ".vui-TextField-AutocompleteResult .item")

        try:
            for button in buttons:
                    if button.text == peca:
                        button.click()
        except:
            print("selecionou peça")
        driver.find_element(By.XPATH ,f'/html/body/div[1]/div/component-dialog-container/component-dialog/div/component-card/div/form/div[1]/div[{elemento_base+1}]/div[1]/input').send_keys(qtdpeca)
        print(qtdpeca)
        driver.find_element(By.XPATH ,f'/html/body/div[1]/div/component-dialog-container/component-dialog/div/component-card/div/form/div[1]/div[{elemento_base+2}]/div[1]/select').click()
        time.sleep(5)
        driver.find_element(By.XPATH ,f'/html/body/div[1]/div/component-dialog-container/component-dialog/div/component-card/div/form/div[1]/div[{elemento_base+2}]/div[1]/select').send_keys(motivopeca)
        
        print(motivopeca)
        driver.find_element(By.XPATH ,f'/html/body/div[1]/div/component-dialog-container/component-dialog/div/component-card/div/form/div[1]/div[{elemento_base+3}]/div[1]/select').send_keys(avaria)
        print(avaria)
        elemento_base +=7
    elemento_base = 44
    pecadiferente = 0
    return pecadiferente
```
